### Solving Linear Equations

$
a_1x + b_1y = d_1
$

$
a_2x + b_2y = d_2
$

becomes

$
\begin{bmatrix} a_1 & b_1\\ a_2 & b_2 \end{bmatrix} \begin{bmatrix} x\\ y \end{bmatrix} = \begin{bmatrix} d_1\\ d_2 \end{bmatrix}

## Optimization
Optimization can be a difficult and time consuming problem to deal with by hand, but by using Scipy's functions it can be greatly simplified. Scipy provides a number of different methods for solving single and multi variable optimization problems with linear and nonlinear constraints. This section will provide some different examples for optimizing based on different functions.

### Single Variable Optimization
The first example will cover single variable, unconstrained optimization. The function we are looking at is y=x^2, which has a single minimum at 0. In order to utilize the Scipy optimization function, the following steps must be taken:


1.   Create a Python function with a single input to model the mathematical function. The function will return the value of the function at the given position.
2.   Create an initial guess for the optimizer with the same dimensions as the input for the previous function.
3.   Input the function and initial_guess into the optimizer.



```python
def function(x):
  return np.power(x,2)

x = np.linspace(-5,5,50)
fig,ax = plt.subplots(1)
ax.plot(x,function(x))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Plot of x^2")
plt.show()
```

```python
initial_guess = [10]
optimize.minimize(function,initial_guess)
```

### Multiple Variable Optimization
Solving a function with multiple variables is very similar to single variables. The Python function must now take in an array containing each variable, and the intial guess must be expanded to match the size of this new input. 

A function called plot_functions is also provided to display the 2d function in the example. The plotter takes as input the function, the bounds, and the solution to the problem and plots them.

The example below will use expand the previous example to 2 dimensions

```python
def plot_function(function, bounds =  [[None,None],[None,None]], solution = None):
  n = 100

  for bound in bounds:
    if bound[0] is None: bound[0] = -10
    if bound[1] is None: bound[1] = 10

  X = np.linspace(bounds[0][0],bounds[0][1],n)
  Y = np.linspace(bounds[1][0],bounds[1][1],n)

  # X,Y = np.meshgrid(X,Y)

  Z1 = np.zeros((n,n))
  for x_index, x in enumerate(X):
    for y_index, y in enumerate(Y):
      Z1[y_index][x_index] = function([x,y])

  print(Z1.shape)

  fig,ax1 = plt.subplots(1,figsize=(10,10))
  cf2 = ax1.contour(X, Y, Z1, 30, linewidths=0.5, colors='k')
  cf2 = ax1.contourf(X, Y, Z1, 50, cmap=plt.cm.rainbow)
  if solution is not None: ax1.scatter(solution[0],solution[1],marker = 'x', color = 'black')
  ax1.set_xlim(bounds[0][0],bounds[0][1])
  ax1.set_ylim(bounds[1][0],bounds[1][1])
  ax1.set_xlabel("x")
  ax1.set_ylabel("y")
  plt.show()
```

```python
def function(pos):
  return np.power(pos[0],2) + np.power(pos[1],2)

initial_guess = [10,10]
solution = optimize.minimize(function,initial_guess)
print(solution)
plot_function(function,solution = solution['x'])
```

### Constrained Optimization
It is common that there are constraints to an optimization problem, and Scipy allows for users to implement both linear and nonlinear constraints. This first example will apply a set of bounds for the previous function.

```python
def function(pos):
  return np.power(pos[0],2) + np.power(pos[1],2)

initial_guess = [10,10]

bnds = [[1,4],[2,3]]  # 1 <= x <= 4, 2 <= y <= 3

solution = optimize.minimize(function,initial_guess,bounds=bnds)
print(solution)
plot_function(function,bounds = bnds,solution = solution['x'])
```

This second example will apply a set of inequality constraints to the problem. Inequality constraints take the form of $Ax-b >= 0$. 

In the example below, the constraint takes the form of $\frac{x+y}{2}-5 >= 0$. This constraint dictates that the sum of the x and y coordinate must be greater then 5. The expected solution then is any point along a circle where x+y = 10.

```python
def function(pos):
  return np.power(pos[0],2) + np.power(pos[1],2)

initial_guess = [10,10]

cons = [{'type':'ineq','fun': lambda pos: (pos[0] + pos[1])/2 - 5}]

solution = optimize.minimize(function,initial_guess,constraints=cons)
print(solution)
plot_function(function,solution = solution['x'])
```

### Local vs Global Optimization
Some problems will have a number of suboptimal local solutions which the algorithm can be trapped in. Scipy's optimize.minimize function is prone to this problem, but the library offers other solutions designed to find the global minimum in these cases.

 The example below will demonstrate this issue, then provide an example of a global optimization method.

```python
def function(pos):
    scalar = 1
    x_trans = 0
    y_trans = 0
    term1 = 0.75 * np.exp(-np.power(9*scalar*pos[0]-2+x_trans,2)/4 - np.power(9*scalar*pos[1]-2+y_trans,2)/4);
    term2 = 0.75 * np.exp(-np.power(9*scalar*pos[0]+1+x_trans,2)/49 - (9*scalar*pos[1]+1+y_trans)/10);
    term3 = 0.5 * np.exp(-np.power(9*scalar*pos[0]-7+x_trans,2)/4 - np.power(9*scalar*pos[1]-3+y_trans,2)/4);
    term4 = -0.2 * np.exp(-np.power(9*scalar*pos[0]-4+x_trans,2) - np.power(9*scalar*pos[1]-7+y_trans,2));
    return -(term1+term2+term3+term4)
```

```python
bnds = [(0,1),(0,1)]
initial_guess = [.8,.4]
solution = optimize.minimize(function,initial_guess,bounds=bnds,method = 'nelder-mead')
print(solution)
plot_function(function,bnds,solution['x'])
```

```python
bnds = [(0,1),(0,1)]
solution = optimize.dual_annealing(function,bounds=bnds)
print(solution)
plot_function(function,bnds,solution['x'])
```

## Interpolation
Interpolation is the estimation of values between known samples. This is useful for datasets that have gaps in their coverage. Scipy provides functions to interpolate data based on a known dataset for one or more dimensions. The first example below demonstrates interpolating data in a dataset recorded from the function $y=e^{\frac{-x}{3}}$

```python
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(0, 10)
y = np.exp(-x/3.0)
f = interpolate.interp1d(x, y)

xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

In the graph, the blue points represent the sampled data, and the orange line represent the interpolated values between the sample. Using the interpolated function, one can generate samples anywhere in the dataset. The example below will use the same function, but remove some of the samples from the dataset.

```python
x = np.arange(0, 10)

x = np.delete(x,(2,3))

y = np.exp(-x/3.0)

f = interpolate.interp1d(x, y)

xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

This second example shows how the interpolated function attempts to estimate values in the gap. Using this same example, we will now show how using a different interpolation function provides a better estimation for the missing data.

```python
x = np.arange(0, 10)

x = np.delete(x,(2,3))

y = np.exp(-x/3.0)

f = interpolate.CubicSpline(x, y)

xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(x, y, 'o', xnew, ynew, '-')
plt.show()
```

The following example will now show how you can interpolate data in multiple dimensions using Scipy. In the example, random samples will be recorded from the function, which will be used to interpolate the values of the signal.

```python
def function(pos):

  if isinstance(pos,list):
    pos = np.array(pos)
    pos = np.expand_dims(pos,axis=0)

  scalar = 1
  x_trans = 0
  y_trans = 0
  term1 = 0.75 * np.exp(-np.power(9*scalar*pos[:,0]-2+x_trans,2)/4 - np.power(9*scalar*pos[:,1]-2+y_trans,2)/4);
  term2 = 0.75 * np.exp(-np.power(9*scalar*pos[:,0]+1+x_trans,2)/49 - (9*scalar*pos[:,1]+1+y_trans)/10);
  term3 = 0.5 * np.exp(-np.power(9*scalar*pos[:,0]-7+x_trans,2)/4 - np.power(9*scalar*pos[:,1]-3+y_trans,2)/4);
  term4 = -0.2 * np.exp(-np.power(9*scalar*pos[:,0]-4+x_trans,2) - np.power(9*scalar*pos[:,1]-7+y_trans,2));
  return -(term1+term2+term3+term4)
```

```python
def plot_original_vs_interpolated_functions(function, interpolation, bounds =  [[None,None],[None,None]], samples = None):
  n = 100

  for bound in bounds:
    if bound[0] is None: bound[0] = -10
    if bound[1] is None: bound[1] = 10

  X = np.linspace(bounds[0][0],bounds[0][1],n)
  Y = np.linspace(bounds[1][0],bounds[1][1],n)

  Z1 = np.zeros((n,n))
  Z2 = np.zeros((n,n))
  interpolated_Z = np.zeros((n,n))

  for x_index, x in enumerate(X):
    for y_index, y in enumerate(Y):
      Z1[y_index][x_index] = function([x,y])

  for x_index, x in enumerate(X):
    for y_index, y in enumerate(Y):
      Z2[y_index][x_index] = interpolation(x,y)

  fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,10))
  cf1 = ax1.contour(X, Y, Z1, 30, linewidths=0.5, colors='k')
  cf1 = ax1.contourf(X, Y, Z1, 50, cmap=plt.cm.rainbow)
  if samples is not None: ax1.scatter(samples[:,0],samples[:,1],marker = 'x', color = 'black')
  ax1.set_xlim(bounds[0][0],bounds[0][1])
  ax1.set_ylim(bounds[1][0],bounds[1][1])
  ax1.set_xlabel("x")
  ax1.set_ylabel("y")
  ax1.set_title("Original Function with Sample Locations")

  cf2 = ax2.contour(X, Y, Z2, 30, linewidths=0.5, colors='k')
  cf2 = ax2.contourf(X, Y, Z2, 50, cmap=plt.cm.rainbow)
  if samples is not None: ax2.scatter(samples[:,0],samples[:,1],marker = 'x', color = 'black')
  ax2.set_xlim(bounds[0][0],bounds[0][1])
  ax2.set_ylim(bounds[1][0],bounds[1][1])
  ax2.set_xlabel("x")
  ax2.set_ylabel("y")
  ax2.set_title("Interpolated Function with Sample Locations")

  plt.show()
```

```python
def generate_samples(function,bounds = [(0,1),(0,1)], number_samples = 100):
  sample_locations = np.empty((number_samples,3))
  sample_locations[:,0] = np.random.uniform(low = bounds[0][0], high = bounds[0][1], size = (sample_locations.shape[0]))
  sample_locations[:,1] = np.random.uniform(low = bounds[1][0], high = bounds[1][1], size = (sample_locations.shape[0]))
  sample_locations[:,2] = function(sample_locations[:,:2])
  return sample_locations
```

```python
bounds = [(0,1),(0,1)]
number_of_samples = 100

samples = generate_samples(function,bounds = bounds, number_samples=number_of_samples)

interpolation = interpolate.Rbf(samples[:,0],samples[:,1],samples[:,2])

plot_original_vs_interpolated_functions(function,interpolation, bounds = bounds, samples = samples)
```

### Retrieving Data with .loc
.loc is used to get data from a DataFrame using it's label. This method can be used to retrieve both columns and rows based on their label. Because the samples of the Iris dataset are unlabled, a sample dataset is created instead to demonstrate .loc.

```python
df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],index=['cobra', 'viper', 'sidewinder'],columns=['max_speed', 'length'])
print(df)
```

```python
print(df.loc["cobra"])
```

```python
print(df.loc[:,"max_speed"])
```

#### Exercise
You will use Numpy and Scipy to perform some analysis on the Iris dataset.
1.   Extract the sepal length and sepal width from the Iris dataset and convert them to a Numpy array
3.   Use Scipy's interpolation.interp1d to interpolate the sepal width based on the sepal lengths.
4.   Plot the values of the sepal width vs. the sepal length with the interpolated function using the provided plotting function.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

def plot_sepal_values(sepal_length, sepal_width, interpolation = None):
  if interpolation is not None: 
    interp_x_values = np.linspace(np.min(sepal_length),np.max(sepal_length),100)
    interp_values = interpolation(interp_x_values)
  fig,ax = plt.subplots(1)
  ax.scatter(sepal_length, sepal_width)
  if interpolation is not None: ax.plot(interp_x_values,interp_values,color = 'r')
  ax.set_xlabel("Sepal Length")
  ax.set_ylabel("Sepal Width")
  ax.set_title("Plot of Sepal Width vs. Sepal Length")
  plt.show()
```

##### Solution

```python
sepal_data = iris.loc[:,["sepal_length","sepal_width"]]
sepal_data = sepal_data.to_numpy()

interpolation = interpolate.interp1d(sepal_data[:,0],sepal_data[:,1])

plot_sepal_values(sepal_data[:,0],sepal_data[:,1],interpolation)
```
